```python
df3=pd.read_csv('sepsis_individual_phenotypes.csv')
df3.head()
```

```python
ax = sns.barplot(data=df3, x='Reported Metadata', y='Clinical Phenotype',hue='Type', palette='mako')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Apart from tissue type and organism, the **least discrepancy** was observed for reporting the **disease condition**, with **100%** being reported on publications and **82.8%** reported on the repositories. On the other hand, the **largest difference** between publications and repositories was observed for the country of residence of the patient. While the country is mentioned on **89.7%** of the publications, it was reported by a mere **3.4%** on public repositories.

```python
df4=pd.read_csv('sepsis_comparison.csv')
df4.head()
```

```python
ax = sns.barplot(data=df4, x='Cohort', y='Reported Metadata', hue='Type', palette='mako')
ax.set_xlabel("Cohort",fontsize=30)
ax.set_ylabel("Reported Metadata",fontsize=30)
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha="right")
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
fig = plt.gcf()
fig.set_size_inches(30, 10)
ax.tick_params(axis='x', labelsize=24)
ax.tick_params(axis='y', labelsize=24)
plt.setp(ax.get_legend().get_texts(), fontsize='24') # for legend text
```

The most reported clinical phenotypes on publications were the **disease condition, organism, and tissue type**, each being **consistently reported** across all studies, while this was also the case for the latter two on public repositories. The **least reported** clinical variable on publications was **ancestry (37.9%)**, while on the repositories it was the **country of residence (3.4%)**.
